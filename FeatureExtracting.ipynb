```python
# read ability
from textstat import flesch_reading_ease, gunning_fog
# Apply Flesch-Kincaid Readability Test
df['flesch_reading_ease'] = df['text'].apply(flesch_reading_ease)

# Apply Gunning Fog Index
df['gunning_fog'] = df['text'].apply(gunning_fog)
```

```python
# Function to calculate average word length
def average_word_length(text):
    words = text.split()
    if len(words) == 0:
        return 0
    return sum(len(word) for word in words) / len(words)

# Apply this function to the 'summary' and 'text' columns
df['avg_word_length_summary'] = df['summary'].apply(average_word_length)
df['avg_word_length_text'] = df['text'].apply(average_word_length)
df
```

```python
from textblob import TextBlob

# Function to calculate subjectivity
def calculate_subjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# Apply this function to the 'summary' and 'text' columns
df['subjectivity_summary'] = df['summary'].apply(calculate_subjectivity)
df['subjectivity_text'] = df['text'].apply(calculate_subjectivity)
df
```

```python
# Diversity of text
def type_token_ratio(text):
    tokens = text.split()
    types = set(tokens)
    if len(tokens) == 0:
        return 0  # Avoid division by zero
    return len(types) / len(tokens)

# Apply this function to the 'text' column
df['ttr_text'] = df['text'].apply(type_token_ratio)
df
```
